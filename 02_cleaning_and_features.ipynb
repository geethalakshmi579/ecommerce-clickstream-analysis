```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
sns.set(style="whitegrid")
%matplotlib inline

print("✅ Libraries imported successfull")
```

```python

# Load the cleaned file we saved in previous notebook
df = pd.read_csv('data/e-shop clothing 2008.csv', sep=';')

print("✅ Data loaded successfully!")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
df_clean = df.copy()
```

```python
df_clean = df.copy()

# Clean column names
df_clean.columns = df_clean.columns.str.strip()
df_clean.columns = df_clean.columns.str.replace(' ', '_', regex=False)
df_clean.columns = df_clean.columns.str.replace('(', '', regex=False)
df_clean.columns = df_clean.columns.str.replace(')', '', regex=False)
df_clean.columns = df_clean.columns.str.lower()

print("Cleaned columns:", df_clean.columns.tolist())
```

```python
# Rename for easier use
df_clean = df_clean.rename(columns={
    'page_1_main_category': 'main_category',
    'page_2_clothing_model': 'clothing_model',
    'session_id': 'session_id',
    'model_photography': 'model_photography'
})

print("\nFinal columns after rename:")
print(df_clean.columns.tolist())
```

```python
# Create useful features
df_clean['date'] = pd.to_datetime(df_clean[['year', 'month', 'day']])
df_clean['day_of_week'] = df_clean['date'].dt.day_name()

category_map = {1: 'Trousers', 2: 'Skirts', 3: 'Blouses', 4: 'Sale'}
df_clean['main_category_name'] = df_clean['main_category'].map(category_map)

print("\n✅ Feature engineering done!")
print(df_clean[['date', 'day_of_week', 'main_category_name']].head())
```

```python
df_clean.to_csv('data/e-shop_cleaned_final.csv', index=False)
print("✅ Final cleaned data saved!")
```
